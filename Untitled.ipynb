```python
from torch.nn.utils import prune
import torch.nn as nn
from torch import tensor
```

```python
class Test(nn.Module):

    def __init__(self):
        super(Test, self).__init__()
        self.hidden = nn.Linear(3, 2)
        self.act = nn.ReLU()
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        return self.output(self.act(self.hidden(x)))
net=Test()
```

```python
print(net.state_dict)
for name,m in net.named_modules():
    print(name)
mask={
    'hidden.weight_mask':tensor([[0,0,0],[1,0,1]]),
    'output.weight_mask':tensor([[0,1]])
}
```

```python
def prune_model_custom_manual(model, mask_dict):
    with torch.no_grad():
        for name,m in model.named_modules():
            if isinstance(m, nn.Conv2d) or isinstance(m, nn.Linear):
                mask_name = name+'.weight_mask'
                if mask_name in mask_dict.keys():
                    eval(f"model.{name}.weight").set_(m.state_dict()['weight']*mask_dict[mask_name])
                else:
                    print('Can not find [{}] in mask_dict'.format(mask_name))
```

```python
prune_model_custom_manual(net,mask)
```

```python
def prune_model_custom(model, mask_dict):

    #print('Pruning with custom mask (all conv layers)')
    for name,m in model.named_modules():
        if isinstance(m, nn.Conv2d) or isinstance(m, nn.Linear):
            mask_name = name+'.weight_mask'
            if mask_name in mask_dict.keys():
                prune.CustomFromMask.apply(m, 'weight', mask=mask_dict[name+'.weight_mask'])
            else:
                print('Can not find [{}] in mask_dict'.format(mask_name))

def prune_model_remove(model):

    #print('Pruning with custom mask (all conv layers)')
    for name,m in model.named_modules():
        if isinstance(m, nn.Conv2d) or isinstance(m, nn.Linear):
            torch.nn.utils.prune.remove(m, name='weight')
            
```

```python
import copy
class Test(nn.Module):

    def __init__(self):
        super(Test, self).__init__()
        self.hidden = nn.Linear(3, 2)
        self.act = nn.ReLU()
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        return self.output(self.act(self.hidden(x)))
net=Test()
mask={
    'hidden.weight_mask':tensor([[0,0,0],[1,0,1]]),
    'output.weight_mask':tensor([[0,1]])
}
print(net.state_dict())
prune_model_custom(net, mask)
print(net.state_dict())
prune_model_remove(net)
print(net.state_dict())
net_save=copy.deepcopy(net)
prune_model_custom_manual(net_save,mask)
print(net_save.state_dict())
```
